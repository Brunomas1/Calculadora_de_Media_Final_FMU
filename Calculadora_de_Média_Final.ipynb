```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

display(HTML('''
<style>
.result-box {
  border: 2px solid #ccc;
  padding: 15px;
  border-radius: 8px;
  background: #f9f9f9;
  color: #000;
  margin-top: 10px;
}
@media (prefers-color-scheme: dark) {
  .result-box {
    border: 2px solid #444;
    background: #333;
    color: #fff;
  }
}
</style>
'''))

a1 = widgets.BoundedFloatText(
    value=0,
    min=0,
    max=10,
    step=0.1,
    description='A1:',
    style={'description_width': '50px'},
)
a2 = widgets.BoundedFloatText(
    value=0,
    min=0,
    max=10,
    step=0.1,
    description='A2:',
    style={'description_width': '50px'},
)
a3 = widgets.BoundedFloatText(
    value=0,
    min=0,
    max=10,
    step=0.1,
    description='A3:',
    style={'description_width': '50px'},
)
a4 = widgets.BoundedFloatText(
    value=0,
    min=0,
    max=10,
    step=0.1,
    description='A4:',
    style={'description_width': '50px'},
)

a5 = widgets.Text(
    value="",
    placeholder="Deixe vazio se não informado",
    description='A5:',
    style={'description_width': '50px'},
)
a6 = widgets.Text(
    value="",
    placeholder="Deixe vazio se não informado",
    description='A6:',
    style={'description_width': '50px'},
)

calc_button = widgets.Button(
    description='Calcular',
    button_style='success'
)

output = widgets.Output()

def calcular_media(b):
    with output:
        clear_output()
        n1 = (a1.value + a2.value + a3.value + a4.value) / 4

        def parse_grade(text):
            text = text.strip().replace(",", ".")
            if text == "":
                return None
            try:
                grade = float(text)
                if grade < 0 or grade > 10:
                    return None
                return grade
            except ValueError:
                return None

        grade_a5 = parse_grade(a5.value)
        grade_a6 = parse_grade(a6.value)

        if grade_a5 is None and grade_a6 is None:
            display(HTML("<p style='color: red;'>É necessário informar pelo menos uma nota para A5 ou A6.</p>"))
            return

        if grade_a5 is not None and grade_a6 is not None:
            exame = max(grade_a5, grade_a6)
            exame_info = f"Utilizando a maior nota entre A5 e A6: {exame:.2f}"
        else:
            exame = grade_a5 if grade_a5 is not None else grade_a6
            exame_info = f"Nota de exame utilizada: {exame:.2f}"

        nota_final = (n1 * 0.4) + (exame * 0.6)

        result_html = f"""
        <div class='result-box'>
            <h2 style="text-align: center;">Resultado da Média Final</h2>
            <p><strong>Média N1 (A1-A4):</strong> {n1:.2f}</p>
            <p><strong>{exame_info}</strong></p>
            <p><strong>Nota Final:</strong> {nota_final:.2f}</p>
        """
        if nota_final >= 6.0:
            result_html += "<h3 style='text-align: center;'>Parabéns! Você foi aprovado na disciplina.</h3>"
        else:
            result_html += "<h3 style='text-align: center;'>Infelizmente, você não atingiu a nota mínima para aprovação.</h3>"
        result_html += "</div>"

        display(HTML(result_html))

calc_button.on_click(calcular_media)

display(
    widgets.VBox([
        widgets.HTML("<h2>Calculadora de Média Final da Disciplina Online</h2>"),
        widgets.HTML("<b>Primeira Etapa: Atividades A1, A2, A3 e A4</b>"),
        a1, a2, a3, a4,
        widgets.HTML("<b>Segunda Etapa: Prova N2 (A5) e Prova Substitutiva (A6)</b>"),
        a5, a6,
        calc_button,
        output
    ])
)
```
